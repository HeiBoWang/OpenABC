```python
import pickle
```

```python
with open('synthID2Vec.pickle','rb') as f:
    synthIDs = pickle.load(f)
```

```python
synthIDs
```

```python
distribution = [[] for j in range(20)]
```

```python
for i in synthIDs.keys():
    arr = synthIDs[i]
    for j in range(len(arr)):
        distribution[j].append(arr[j])
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(20):
    plt.hist(distribution[i])
    plt.title('Synthesis transformation distribution (Step: '+str(i+1)+')')
    plt.xlabel('Synthesis Transformation')
    plt.ylabel('# Counts')
    plt.savefig('step'+str(i+1)+'.png',fmt = 'png',bbox_inches='tight')
    plt.show()
```
